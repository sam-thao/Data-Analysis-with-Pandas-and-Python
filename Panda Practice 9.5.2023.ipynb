# Creating New column from Existing Column

```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
nba['Age in a Decade'] = nba['Age'] + 10 #right side of = is evaluated first
```

```python
nba.head()
```

```python
nba['Age in a Decade'] = nba['Age'].add(10)
```

```python
nba.head()
```

```python
nba['New Salary'] = nba['Salary'] - 5000000
```

```python
nba.head()
```

```python
nba['Weight'] * 5
nba['Weight in kilograms'] = nba['Weight'].mul(.4535)
```

```python
nba.head()
```

# Drop Rows with Null Values

```python
nba = pd.read_csv('nba.csv')
nba.tail()
```

```python
nba.dropna() #remove rows with missing values
nba.dropna(how = 'all')
```

```python
nba.dropna(subset = ['College'])
```
